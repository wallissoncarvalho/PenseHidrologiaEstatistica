# Análise Exploratória de Dados

```python
# Import numpy
import numpy as np

# Set default Seaborn style
import seaborn as sns
sns.set()

# Compute number of data points: n_data
n_data=len(vazoes['16100000'].dropna())

# Number of bins is the square root of number of data points: n_bins
n_bins=np.sqrt(n_data)

# Convert number of bins to integer: n_bins
n_bins=int(n_bins)

# Plot the histogram
_ = plt.hist(vazoes['16100000'],bins=n_bins)

# Label axes
_ = plt.xlabel('Vazoes (m³/s)')
_ = plt.ylabel('count')

# Show histogram
plt.show()
```

## Função de Distribuição Acumulada Empírica (ECDF)

```python
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1,n+1) / n

    return x, y
```

```python
sns.set()
# Compute ECDF for versicolor data: x_vers, y_vers
x_1, y_1 = ecdf(vazoes['16100000'])
x_2, y_2 = ecdf(vazoes['16080000'])

# Generate plot
_ = plt.plot(x_1, y_1, marker='.', linestyle = 'none')
_ = plt.plot(x_2, y_2, marker='.', linestyle = 'none')

# Label the axes
plt.legend(('16100000', '16080000'), loc='lower right')
_ = plt.xlabel('Vazão (m³/s)')
_ = plt.ylabel('ECDF')

# Display the plot
plt.margins(0.02)
plt.show()
```

```python
# Compute the covariance matrix: covariance_matrix
df=vazoes.dropna(subset=['16080000','16100000'])
covariance_matrix=np.cov(df['16100000'].dropna(),df['16080000'].dropna())

# Print covariance matrix
print(covariance_matrix)

# Extract covariance of flows of stations: flows_cov
flows_cov=covariance_matrix[0,1]

# Print the flows stations covariance
print(flows_cov)
```

```python
def pearson_r(x, y):
    """Compute Pearson correlation coefficient between two arrays."""
    # Compute correlation matrix: corr_mat
    corr_mat=np.corrcoef(x,y)

    # Return entry [0,1]
    return corr_mat[0,1]

# Compute Pearson correlation coefficient: r
r=pearson_r(df['16100000'].dropna(),df['16080000'])

# Print the result
print(r)
```
