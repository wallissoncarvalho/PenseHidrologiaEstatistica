# Análise de Séries Temporais

```python
from dados_hidroweb import converte_dados
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
from graficos import boxplot_mes,prec_mensal

dados=converte_dados('Dados')
data_construcao=pd.to_datetime(['31/12/1984'])
dados=dados.loc[:data_construcao[0]]
chuvas=dados[['00160000','00259000']]
vazoes=dados[['16100000','16080000']]
#vazoes=vazoes.dropna(subset=['16080000','16100000',])
#16100000 - Jusante ('Vazão')
#16080000 - Montante('Vazão')
#00160000 - Chuva
#00259000 - Chuva
```

```python
boxplot_mes(vazoes,1,2,(10,5))
```

## Médias Moveis

```python
dados_chuva = pd.Series(dados['00160000'].groupby(pd.Grouper(freq='M')).sum(), name='Prec')
dados_chuva2 = pd.Series(dados['00259000'].groupby(pd.Grouper(freq='M')).sum(), name='Prec2')
dados_vazao_jus = pd.Series(dados['16100000'].groupby(pd.Grouper(freq='M')).mean(), name='Jusante') 
dados_vazao_mont = pd.Series(dados['16080000'].groupby(pd.Grouper(freq='M')).mean(), name='Montante')
dados_month = pd.DataFrame([dados_vazao_jus, dados_vazao_mont, dados_chuva,dados_chuva2]).T
dados_month=dados_month.dropna(subset=['Prec','Jusante','Montante','Prec2'])
dados_month.head()
```

```python
chuvas['Media_3']=chuvas['00259000'].rolling(window = 3).mean()
chuvas['Media_7']=chuvas['00259000'].rolling(window = 7).mean()
chuvas['Media_15']=chuvas['00259000'].rolling(window = 15).mean()
```

```python
chuvas['00259000'].plot(legend=True)
chuvas['Media_3'].plot(legend=True)
chuvas['Media_7'].plot(legend=True)
chuvas['Media_15'].plot(legend=True)
```

## Correlação serial

```python
import thinkstats2
import thinkplot
corr_serial_jusante = thinkstats2.SerialCorr(vazoes['16100000'].dropna(), lag=1)
corr_serial_montante = thinkstats2.SerialCorr(vazoes['16080000'].dropna(), lag=1)
print('Correlação serial da Série de Vazões à Jusante: {} \nCorrelação serial da Série de Vazões à Montante: {}'
      .format(round(corr_serial_jusante,3),round(corr_serial_montante,3)))
```

## Auto Correlação

```python
import statsmodels.tsa.stattools as smtsa

aut_corr_jusante = smtsa.acf(vazoes['16100000'].dropna(), nlags=365, unbiased=True)
day = np.arange(0,366,1)
```

```python
plt.plot(day,aut_corr_jusante)
```
