```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
import seaborn as sns
import calendar
import scipy.stats as ss

chuvas = pd.read_excel('Chuvas_Paraiba.xls',index_col='Data')
chuvas.dropna(inplace=True)
```

```python
def ano_hidrologico(selecionados):
    media_mensal=selecionados.groupby(selecionados.index.month).mean()
    mes_menor_media=calendar.month_abbr[media_mensal.idxmin()[0]+1].upper()
    frequencia='AS-'+mes_menor_media
    acumulado_anual=chuvas.groupby(pd.Grouper(freq=frequencia)).sum().sort_index()
    acumulado_anual.head()
    return acumulado_anual



def normal_probability_plot(acumulado_anual):
    sorted_annual_mean=acumulado_anual.sort_values()
    standard_normal=pd.Series(np.random.normal(size=len(acumulado_anual))).sort_values()
    model_normal=ss.norm.pdf(standard_normal, acumulado_anual.mean(), acumulado_anual.std())
    model_normal=pd.Series(model_normal)
    model_normal=model_normal.sort_values()
    a1=plt.plot(standard_normal.values,sorted_annual_mean.values)
    #a2=plt.plot(standard_normal.values,model_normal.values)
    #a1.legend(('empírico','analítico'))
    
    

    
def normal_distribution(acumulado_anual):
    ax = acumulado_anual.plot(kind='density')
    mean = acumulado_anual.mean()
    std_dev = acumulado_anual.std()
    lin = np.linspace(0, acumulado_anual.max(), num=50)
    normal = ss.norm.pdf(lin, mean, std_dev)
    plt.plot(lin, normal)
    ax.legend(('empírico','analítico'))

acumulado_anual=ano_hidrologico(chuvas)
```

```python
normal_probability_plot(acumulado_anual['00936010'])
```

```python
normal_distribution(acumulado_anual['00936010']) 
```

```python
a=pd.Series(acumulado_anual[acumulado_anual['00936010']>1600]['00936010'].index)
a.sort_values()
diffs=[]
for i in range(len(a)-1):
    diffs.append(((a[i+1]-a[i]).days/365))
```

```python
def expon_distribution(diff_events):
    ax = diff_events.plot(kind='density')
    
    #cumulative=np.cumsum(diff_events)
    #cumulative_prob=cumulative/pd.Series(cumulative).max()
    #plt.plot(diff_events.sort_values(),cumulative_prob)
    
    mu,sigma=ss.expon.fit(diff_events)
    lin = np.linspace(0, diff_events.max(), num=50)
    expon = ss.expon.pdf(lin, mu, sigma)
    plt.plot(lin, expon)
    
    expon_cdf = ss.expon.cdf(lin, mu, sigma)
    plt.plot(lin, expon_cdf)
    #ax.legend(('Empírico PDF','Empírico CDF','Analíico PDF','Analítico CDF'))
    ax.legend(('Empírico PDF','Analíico PDF','Analítico CDF'))
    plt.xlim(1)
    
expon_distribution(pd.Series(diffs))
```

```python
def log_normal_probability_plot(acumulado_anual):
    sorted_annual_mean=acumulado_anual.apply(np.log).sort_values()
    standard_normal=pd.Series(np.random.normal(size=len(acumulado_anual))).sort_values()
    plt.plot(standard_normal.values,sorted_annual_mean.values)
    
log_normal_probability_plot(acumulado_anual['00936010'])
```

```python
def ano_hidrologico_max(selecionados):
    media_mensal=selecionados.groupby(selecionados.index.month).max()
    mes_menor_media=calendar.month_abbr[media_mensal.idxmin()[0]+1].upper()
    frequencia='AS-'+mes_menor_media
    acumulado_anual=chuvas.groupby(pd.Grouper(freq=frequencia)).sum().sort_index()
    acumulado_anual.head()
    return acumulado_anual

maximos_anuais=ano_hidrologico_max(chuvas)
maximos_anuais=maximos_anuais[maximos_anuais['00936010']!=0]
```

```python
maximos_anuais=maximos_anuais['00936010']
a,b,c=ss.pareto.fit(maximos_anuais)
```

```python
def pareto_distribution(maximos_anuais):
    #ax = maximos_anuais.plot(kind='density')
    
    a,b,c=ss.pareto.fit(maximos_anuais)
    lin = np.linspace(maximos_anuais.min(), maximos_anuais.max(), num=50)
    pareto = ss.pareto.pdf(lin, a,b,c)
    fig, ax1 = plt.subplots()
    pareto_cdf = ss.pareto.cdf(lin, a,b,c)
    ax2 = ax1.twinx()
    ax1.plot(lin, pareto, color='red')
    ax2.plot(lin, pareto_cdf, color='blue')
    
    

pareto_distribution(maximos_anuais)
```
