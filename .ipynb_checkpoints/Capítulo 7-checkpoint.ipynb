### Correlação e Covariância

#### Coeficiente de Correlação de Pearson (p)

• É a razão entre a Covariância de duas variáveis (X,Y) pelo produto do desvio padrão de X e de Y.

•  A correlação de Pearson é sempre entre -1 e +1. Se ρ é positivo, dizemos que a correlação é positiva. Se ρ é negativo, a correlação é negativa.

• A magnitude de ρ indica a força da correlação. Se ρ é 1 ou -1, as variáveis são perfeitamente correlacionadas, o que significa que se você conhece uma, você pode fazer uma previsão perfeita sobre o outro.

• Se a correlação de Pearson estiver próxima de 0, é tentador concluir que não há relação entre as variáveis, mas essa conclusão não é válida. A correlação de Pearson mede apenas relacionamentos lineares.

• Usando numpy.cov - Retorna uma matriz 2x2 com elementos iguais a:

                            cov(a,a)  cov(a,b)

                            cov(a,b)  cov(b,b)

```python
cov=np.cov(df['16080000'],df['16100000'])
Sx=df['16080000'].std()
Sy=df['16100000'].std()
cov[0][1]
p=cov[0][1]/(Sx*Sy)

print('Covariância = {} \nDesvio padrão da estação 16080000 = {}\n'
      'Desvio padrão da estação 16100000 = {}\nPearson Coef. = {}'
     .format(round(cov[0][1],2),round(Sx,2),round(Sy,2),round(p,2)))
```

```python
df.corr(method='pearson')['16080000']['16100000'].round(2)
```

• O coeficiente de correlação de Pearman funciona bem se a relação entre as variáveis é linear e sse as variáveis são aproximadamente normais. Mas, não é robusto na presença de outliers. 

#### Correlação de ordem de classificação de Spearman

• É uma alternatica para mitidar o efeito de outliers e distribuições enviesadas.

```python
xs = df['16080000'].rank()
```

```python
df.corr(method='spearman')
```

```python
df.corr(method='spearman')['16080000']['16100000'].round(2)
```
