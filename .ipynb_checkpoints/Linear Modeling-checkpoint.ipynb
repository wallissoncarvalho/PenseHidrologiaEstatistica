## Introdução

```python
from dados_hidroweb import converte_dados
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
vazoes=converte_dados('Dados')
data_construcao=pd.to_datetime(['31/12/1984'])
vazoes=vazoes.loc[:data_construcao[0]]
vazoes=vazoes.dropna(subset=['16080000','16100000'])
x = vazoes['16080000']
y = vazoes['16100000']
```

```python
def pearson_r(x, y):
    """Compute Pearson correlation coefficient between two arrays."""
    # Compute correlation matrix: corr_mat
    corr_mat=np.corrcoef(x,y)

    # Return entry [0,1]
    return corr_mat[0,1]

# Compute Pearson correlation coefficient: r
r=pearson_r(x,y)

# Print the result
print ("O coeficiente de correlação de person é: r={}".format(round(r,2)))

# Plot the illiteracy rate versus fertility
_ = plt.plot(x, y, marker='.', linestyle='none')

# Set the margins and label axes
plt.margins(0.02)
_ = plt.xlabel('16080000')
_ = plt.ylabel('16100000')

# Show the plot
plt.show()
```

```python
# Perform linear regression: a, b
a, b = np.polyfit(x,y,1)

# Print the slope and intercept
print ('Parâmetro a ={}, e b = {}'.format(round(a,2),round(b,2)))

# Generate theoretical x and y data: x_theor, y_theor
x_theor = np.array([x.min(), x.max()])
y_theor = x_theor * a + b

# Plot the Anscombe data and theoretical line
_ = plt.plot(x, y, marker='.', linestyle='none')
_ = plt.plot(x_theor, y_theor, marker=' ', linestyle='-')

# Label the axes
plt.xlabel('x')
plt.ylabel('y')

# Show the plot
plt.show()
```

```python
# Specify slopes to consider: a_vals
a_vals = np.linspace(0,2.3,200)

# Initialize sum of square of residuals: rss
rss = np.empty_like(a_vals)

# Compute sum of square of residuals for each value of a_vals
for i, a in enumerate(a_vals):
    rss[i] = np.sum((y - a*x - b)**2)

# Compute the values of min
rss_minimo = rss.min()
loc_result = np.where(rss == rss.min())[0]
a_min = a_vals[loc_result]
# Plot the RSS
_ = plt.plot(a_vals, rss, '-')
_ = plt.plot(a_min,rss_minimo, '.',color='red',markersize=12)

#Plot the min

plt.xlabel('Coeficiente A (X / Y)')
plt.ylabel('RSS')

plt.show()
```

```python
import thinkstats2
import thinkplot
def SamplingDistributions(x,y, iters=101):
    t = []
    for _ in range(iters):
        sample = thinkstats2.ResampleRows(vazoes)
        x = sample["16080x000"]
        y = sample["16100000"]
        estimates = thinkstats2.LeastSquares(x, y)
        t.append(estimates)
    inters, slopes = zip(*t)
    return inters, slopes
```

```python
inters, slopes = SamplingDistributions(x, y, iters=1001)
```

```python
def Summarize(estimates, actual=None):
    mean = thinkstats2.Mean(estimates)
    stderr = thinkstats2.Std(estimates, mu=actual)
    cdf = thinkstats2.Cdf(estimates)
    ci = cdf.ConfidenceInterval(90)
    print('mean: ', mean,  
          '\nSE: ',stderr, 
          '\nCI: ', ci)
```

```python
Summarize(inters)
```

```python
Summarize(slopes)
```

```python
def PlotConfidenceIntervals(xs, inters, slopes, percent=90, **options):
    fys_seq = []
    for inter, slope in zip(inters, slopes):
        fxs, fys = thinkstats2.FitLine(xs, inter, slope)
        fys_seq.append(fys)

    p = (100 - percent) / 2
    percents = p, 100 - p
    low, high = thinkstats2.PercentileRows(fys_seq, percents)
    thinkplot.FillBetween(fxs, low, high, **options)
```

```python
PlotConfidenceIntervals(x, inters, slopes, percent=90, 
                        color='gray', alpha=0.3, label='90% CI')
PlotConfidenceIntervals(x, inters, slopes, percent=50,
                        color='gray', alpha=0.5, label='50% CI')

thinkplot.Config(xlabel="16080000", ylabel='16100000', xlim=[0, 1900])
```
